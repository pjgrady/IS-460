```python
df = pd.read_csv(path + filename, usecols = ['CallDateTime'], nrows = 8, skiprows = lambda i: i % 1000000 != 0)
```

```python
df
```

```python
df.dtypes
```

```python
df['CallDateTime'] = pd.to_datetime(df['CallDateTime'], format = '%m/%d/%Y %I:%M:%S %p')
```

```python
df.dtypes
```

```python
df['Year'] = df['CallDateTime'].dt.year
```

```python
df
```

```python
df['Quarter'] = df['CallDateTime'].dt.quarter
```

```python
df
```

```python
df['Month'] = df['CallDateTime'].dt.month
```

```python
df
```

```python
df['Day'] = df['CallDateTime'].dt.day
df['Hour'] = df['CallDateTime'].dt.hour
```

```python
df
```

```python
df['DayOfTheWeek'] = df['CallDateTime'].dt.dayofweek
```

```python
df
```

```python
df['DayOfTheYear'] = df['CallDateTime'].dt.dayofyear
```

```python
df
```

```python
df['MonthName'] = df['CallDateTime'].dt.strftime('%B')
df['MonthNameAbbrev'] = df['CallDateTime'].dt.strftime('%b')
```

```python
df
```

```python
df['DayName'] = df['CallDateTime'].dt.strftime('%A')
df['DayNameAbrev'] = df['CallDateTime'].dt.strftime('%a')
```
